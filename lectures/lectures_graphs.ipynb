```python
import numpy as np
import matplotlib.pyplot as plt
```

## Lecture 07b
### Activation functions

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-4, 4, 100)
ReLU = np.maximum(0, x)
sgm_values = sigmoid(x)
GeLU = x * sigmoid(1.702 * x)

plt.plot(x, sgm_values, label='sigmoid')
plt.plot(x, ReLU, label='ReLU')
plt.plot(x, GeLU, label='GeLU')
plt.xlabel('x')
plt.ylabel('Activation value')
plt.title('Activation functions')
plt.legend()
plt.show()
```
